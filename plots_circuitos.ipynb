# Plots circuitos

Circuitos mixers

```python
num_qubits = 10
```

```python
from qiskit.circuit.library import PauliTwoDesign, TwoLocal, EfficientSU2, RealAmplitudes
from qiskit import QuantumCircuit

def create_ring_mixer_display():
    ring_mixer = QuantumCircuit(num_qubits)
    for i in range(0, num_qubits, 2):
        ring_mixer.x(i)
    return ring_mixer

def create_parity_ring_mixer_display():
    parity_ring_mixer = QuantumCircuit(num_qubits)
    for i in range(0, num_qubits, 2):
        parity_ring_mixer.x(i)
    parity_ring_mixer.x(num_qubits - 1)
    return parity_ring_mixer

def create_full_mixer_display():
    full_mixer = QuantumCircuit(num_qubits)
    for qubit in range(num_qubits):
        full_mixer.x(qubit)
    return full_mixer
```

```python
create_ring_mixer_display().draw('mpl')
```

```python
create_parity_ring_mixer_display().draw('mpl')
```

```python
create_full_mixer_display().draw('mpl')
```

Circuitos ansatzes

```python
EfficientSU2(num_qubits).decompose().draw('mpl')
```

```python
RealAmplitudes(num_qubits).decompose().draw('mpl')
```

```python
PauliTwoDesign(num_qubits, reps=1).decompose().draw('mpl')
```

```python
TwoLocal(num_qubits, "ry", "cz", reps=1, entanglement="full").decompose().draw('mpl')
```
